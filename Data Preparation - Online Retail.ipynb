```python
#Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Loading Data
df = pd.read_excel("/Online Retail.xlsx")
```

```python
#Checking Data

df.info()
df.isnull().sum()
```

```python
#Checking and Solving NULL Value

df[df['Description'].isnull() & df['CustomerID'].isnull()]

null_value = df['Description'].isnull() & df['CustomerID'].isnull()
thats_null = ~null_value
df = df[thats_null]
```

```python
df.describe()
```

```python
#Sorting data from anomalies value

df = df[df['UnitPrice'] > 0]
df = df[df['Quantity'] > 0]
```
